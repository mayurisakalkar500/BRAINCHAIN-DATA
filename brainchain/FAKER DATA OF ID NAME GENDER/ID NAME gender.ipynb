```python
import pandas as pd
import numpy as np
from faker import Faker
```

```python
fake = Faker()
```

```python
# number of records
num_records = 10000
```

```python
# Define the distribution for age
average_age = 40
age_std_dev = 10
```

```python
# Define the probabilities for gender
gender_choices = ['Male', 'Female']
gender_probs = [0.7, 0.3]
```

```python
# Generate the data
data = {
    "ProfileID": list(range(1, num_records + 1)),  # Auto-incrementing ProfileID
    "UserID": list(range(1, num_records + 1)),  # Assuming UserID is the same as ProfileID for simplicity
    "FirstName": [fake.first_name() for _ in range(num_records)],
    "LastName": [fake.last_name() for _ in range(num_records)],
    "Age": np.random.normal(average_age, age_std_dev, num_records).astype(int),
    "Gender": np.random.choice(gender_choices, num_records, p=gender_probs),
    "Bio": [fake.text(max_nb_chars=200) for _ in range(num_records)]  # Generating a Bio with max 200 characters
}
```

```python
# Convert to DataFrame
df = pd.DataFrame(data)
```

```python
# Save the DataFrame to a CSV file
file_path = 'fake_profile_data.csv'
df.to_csv(file_path, index=False)
```

```python
print(f"Data saved to {file_path}")
```
